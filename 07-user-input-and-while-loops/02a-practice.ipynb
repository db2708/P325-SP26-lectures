# User Input and While Loops

## Practice

When buying tickets to a movie, your age determines which movies you're allowed to see (G, PG, PG-13, R). Write a `while` loop that asks a user for their age, then prints the highest rated movie they are allowed to see. Include a quit prompt that will use `break` to stop the loop. Can you think of a reasonable use for the `continue` statement in this example?

```python

active = True

while active:
    response = input("Enter your age:")

    if response == 'q' or 'quit' or 'Q' or 'Quit' or 'QUIT':
        break 

    age = float(response)
    
    if age < 18:
        print("You are a minor, you can see movies rated G, PG, and PG-13.")
    elif age >= 18:
        print("You are an adult, you can see movies of any rating.")
    elif age < 13: 
        print("You are a child, you can see movies rated G and PG.")
    elif age < 8:
        print('You are a young child, you can see movies rated G.')
```

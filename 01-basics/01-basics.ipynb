# Running Python programs

In this class, we'll discuss three of the main ways to run python programs:
1. Using the Python prompt (aka, calculator mode)
2. Writing a python program (file ending in `.py`)
3. Writing an IPython (or Jupyter) notebook (file ending in `.ipynb`)

## Python Prompt

The Python prompt is generally accessed by typing `python3` into a terminal. For example, after opening a terminal on Ubuntu using CTRL+ALT+T, you could start a Python prompt like we did in class yesterday:
```bash
$ python3
Python 3.12.3 (main, Nov 6 2025, 13:44:16) [GCC 13.3.0] on linux
Type “help”, ”copyright”, “credits”, or “license” for more information.
>>>
```
Remember, in this class `$` indicates a terminal prompt, and `>>>` indicates a Python prompt.

You can also open a terminal in VS Code using the `Terminal` > `New Terminal` menu. Give it a try! When you "press play" to run your code, this also opens a terminal, with the additional step of running your program.

Once in the Python prompt, you can enter any valid Python code and it will be interpreted line by line. For example:
```python
>>> print("Hello World!")
Hello World!
```
This makes the Python prompt an excellent calculator. (More on this in a later section.)

## Python Program

Python commands can also be collected in a single file, where the extension `.py` indicates that the file is a Python program.
The contents of the file are run through the Python interpreter, which reads through the program and determines what each word in the program means. For example, when the interpreter sees the word `print` followed by parentheses, it prints to the screen whatever is inside the parentheses.

As you write your program, your editor highlights the different parts of your program in different ways. For example, it recognizes that `print()` is the name of a function and displays that word in one color. (Yellow, for me.) It recognizes that `"Hello World!"` is not Python code, and displays that phrase in a different color. This is called syntax highlighting and is incredibly useful for beginners and experts.

You've already made one program: the Hello World program you wrote on the first day to test that everything is working. Congratulations, you are officially a programmer!

## IPython (or Jupyter) Notebook

The final way to run Python code that we'll talk about is in IPython Notebooks, now known as Jupyter Notebooks. It is an interactive, web-based application that allows you to create and share computational documents that combine live code, explanatory text, mathematical equations, and images in one document.

For example, this document is a Jupyter notebook. So far, only markdown cells have been used. However, you can also include live code, such as:

````python
print("Hello Jupyter!")
````

You can execute the above code by selecting the cell and pressing SHIFT+ENTER. You can also change the code in the cell and execute it again. Update the above cell and give it a try! In general, you can execute any cell in a Jupyter notebook in any order you want. However, if one cell depends on code defined in another cell, it will print an error unless the other cell is executed first.

When executing the above code, you were likely asked to select a kernel. This is how you tell the notebook which version of Python to use.
